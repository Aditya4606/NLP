```python
def is_valid(word: str) -> bool:
    q = "q0"
    for ch in word:
        if q == "q0":
            if 'a' <= ch <= 'z':
                q = "q1"
            else:
                q = "q_dead"
        elif q == "q1":
            if 'a' <= ch <= 'z':
                q = "q1"
            else:
                q = "q_dead"
        else:  # q_dead
            q = "q_dead"

    return q == "q1"
```

```python
tests = ["cat", "dog", "a", "zebra", "dog1", "1dog", "DogHouse", "Dog_house", " cats", ""]
for t in tests:
    print(f"'{t}':", "Accepted" if is_valid(t) else "Not Accepted")
```
